Approaches to Create An Object of Ndarray of Numpy

```python
#Array
l1=[10,20,30,40,50,79]
print(l1,type(l1))
a=np.array(l1)
print(a,type(a))
```

```python
#Attributes of Array
t=np.array([10,20,30,40,50,60])
print(t)
print("Dimension of Array",t.ndim)
print("Data Type of Array",t.dtype)
print("Size of Array",t.size)
print("Shape of Array",t.shape)
b=t.reshape(2,3)
print(b)
```
